```
import spacy
nlp = spacy.load('en_core_web_md')
```

```
from spacy.matcher import Matcher
```

```
matcher = Matcher(nlp.vocab)
patterns = [
    {
        'TEXT': {'REGEX': '^@'}
        
    }
]
matcher.add('UserName', [patterns])
```

```
def helper(text):
    doc = nlp(text)
    matches = matcher(doc)
    matches.sort(key= lambda x: x[1])
    temp = []
    for m in matches:
        temp.append(doc[m[1]:m[2]])
    return temp    
```

```
import pandas as pd
```

```
df = pd.read_csv('cleand_tweets.csv')
un = df['tweet'].apply(lambda x: helper(x))
```

```
user_dict = {}

for index, users in un.items():
    if users:
        for user in users:

            if str(user) in user_dict:
                user_dict[str(user)] += 1
            else:
                user_dict[str(user)] = 1
         
user_series = pd.Series(user_dict)
```
